```python
date_cols = ["dt_opened","upload_dt","opened_dt","last_paymt_dt","closed_dt","reporting_dt"]

for col in date_cols:
    data1[col] = pd.to_datetime(data1[col], errors="coerce")
```

```python
data1["opened_dt"].fillna(data1["dt_opened"], inplace=True)
```

Some missing values are meaningful.
For example, if last_paymt_dt is missing, it may mean the customer never made a payment.

If I directly fill missing values, the model cannot tell which values were originally missing. That removes useful information.

So I did two steps:

Create a flag column (never_paid_flag) that stores whether the value was missing.

Fill missing values with a logical replacement (opened_dt) so calculations can still be done.

This way:

The dataset has no missing values (good for models)

The information that the value was missing is still preserved

So the model can learn patterns like:
customers who never paid → higher risk.

```python
data1["never_paid_flag"] = data1["last_paymt_dt"].isnull().astype(int)
data1["last_paymt_dt"].fillna(data1["opened_dt"], inplace=True)
```

```python
data1["is_active_account"] = data1["closed_dt"].isnull().astype(int)
data1["closed_dt"].fillna(data1["reporting_dt"], inplace=True)
```

```python
data1["high_credit_amt"].fillna(0, inplace=True) # hign credit amount means Highest amount of credit customer ever used in that account
```

```python
data1["amt_past_due"].fillna(0, inplace=True)
#Amount customer has not paid yet but should have paid.
#Because logically missing past due = customer has no overdue So replacing with 0 is correct.
```

Missing values in creditlimit usually mean the account does not have a credit facility, not that the value is unknown.
So we fill missing values with 0 to represent “no credit limit”.

We also create a flag column (has_creditlimit) before filling so we don’t lose the original information about which values were missing.

This helps the model understand:

whether customer had credit access

how much credit they had

So we preserve information and keep data usable for modeling.

```python
data1["has_creditlimit"] = data1["creditlimit"].notnull().astype(int)
data1["creditlimit"].fillna(0, inplace=True)
```

```python
data1["has_cashlimit"] = data1["cashlimit"].notnull().astype(int)
data1["cashlimit"].fillna(0, inplace=True)
```

```python
data1["has_interest"] = data1["rateofinterest"].notnull().astype(int)
data1["rateofinterest"] = pd.to_numeric(data1["rateofinterest"], errors="coerce")
data1["rateofinterest"].fillna(0, inplace=True)
```

```python
data1["has_paymentfreq"] = data1["paymentfrequency"].notnull().astype(int)
data1["paymentfrequency"].fillna(0, inplace=True)
```

```python
data1["has_payment"] = data1["actualpaymentamount"].notnull().astype(int)
data1["actualpaymentamount"].fillna(0, inplace=True)
```

```python
data2=pd.read_csv("Cust_Demographics.csv")
```

```python
data2.head()
```

```python
missing = data2.isnull().mean()*100
missing.sort_values(ascending=False)
```

```python
data2.info()
```

```python
data2["dt_opened"] = pd.to_datetime(data2["dt_opened"], errors="coerce")
data2["entry_time"] = pd.to_datetime(data2["entry_time"], errors="coerce")
```

```python
missing = data2.isnull().mean()*100
drop_cols = missing[missing > 60].index
data2.drop(columns=drop_cols, inplace=True)
```

```python
const_cols = [col for col in data2.columns if data2[col].nunique() <= 1]
data2.drop(columns=const_cols, inplace=True)
```

```python
data2.info()
```

```python
# -------- NUMERIC COLUMNS --------
num_cols = data2.select_dtypes(include=["float64","int64"]).columns
data2[num_cols] = data2[num_cols].fillna(data2[num_cols].median())
```

```python
# -------- CATEGORICAL COLUMNS --------
cat_cols = data2.select_dtypes(include="object").columns
data2[cat_cols] = data2[cat_cols].fillna("Missing")
```

```python
# -------- DATE COLUMNS --------
date_cols = data2.select_dtypes(include="datetime64[ns]").columns
data2[date_cols] = data2[date_cols].fillna(pd.Timestamp("1900-01-01"))
```

```python
data3=pd.read_csv("Cust_Enquiry.csv")
```

```python
data3.head()
```

```python
data3.info()
```

```python
data3["dt_opened"] = pd.to_datetime(data3["dt_opened"], errors="coerce")
data3["upload_dt"] = pd.to_datetime(data3["upload_dt"], errors="coerce")
data3["enquiry_dt"] = pd.to_datetime(data3["enquiry_dt"], errors="coerce")
```

```python
num_cols = data3.select_dtypes(include=["float64","int64"]).columns
data3[num_cols] = data3[num_cols].fillna(data3[num_cols].median())

cat_cols = data3.select_dtypes(include="object").columns
data3[cat_cols] = data3[cat_cols].fillna("Missing")
```

```python
data3.info()
```

```python
enq= data3.groupby("customer_no").agg(
    total_enquiries=("enquiry_dt","count"),
    avg_enq_amt=("enq_amt","mean"),
    max_enq_amt=("enq_amt","max"),
    last_enq_date=("enquiry_dt","max")
).reset_index()
```

```python
enq.head()
```

```python
enq.info()
```

```python
acc = data1.groupby("customer_no").agg(  #aggregating account table
    total_accounts=("acct_type","count"),
    total_balance=("cur_balance_amt","sum"),
    avg_balance=("cur_balance_amt","mean"),
    total_past_due=("amt_past_due","sum"),
    max_credit=("high_credit_amt","max"),
    avg_credit=("high_credit_amt","mean")
).reset_index()
```

```python
df = data2.merge(acc, on="customer_no", how="left")
```

```python
df = df.merge(enq, on="customer_no", how="left")
```

```python
df.head()
```

```python
df.info()
```

```python
 today = pd.Timestamp.today()#Dates cannot go into model directly.

df["account_age_days"] = (today - df["dt_opened"]).dt.days
df["customer_tenure_days"] = (today - df["entry_time"]).dt.days
df["days_since_last_enq"] = (today - df["last_enq_date"]).dt.days
```

```python
df["days_since_last_enq"].fillna(9999, inplace=True)
```
