```python
data1["opened_dt"].fillna(data1["dt_opened"], inplace=True)
```

Some missing values are meaningful.
For example, if last_paymt_dt is missing, it may mean the customer never made a payment.

If I directly fill missing values, the model cannot tell which values were originally missing. That removes useful information.

So I did two steps:

Create a flag column (never_paid_flag) that stores whether the value was missing.

Fill missing values with a logical replacement (opened_dt) so calculations can still be done.

This way:

The dataset has no missing values (good for models)

The information that the value was missing is still preserved

So the model can learn patterns like:
customers who never paid → higher risk.

```python
data1["never_paid_flag"] = data1["last_paymt_dt"].isnull().astype(int)
data1["last_paymt_dt"].fillna(data1["opened_dt"], inplace=True)
```

```python
data1["is_active_account"] = data1["closed_dt"].isnull().astype(int)
data1["closed_dt"].fillna(data1["reporting_dt"], inplace=True)
```

```python
data1["high_credit_amt"].fillna(0, inplace=True) # hign credit amount means Highest amount of credit customer ever used in that account
```

```python
data1["amt_past_due"].fillna(0, inplace=True)
#Amount customer has not paid yet but should have paid.
#Because logically missing past due = customer has no overdue So replacing with 0 is correct.
```

Missing values in creditlimit usually mean the account does not have a credit facility, not that the value is unknown.
So we fill missing values with 0 to represent “no credit limit”.

We also create a flag column (has_creditlimit) before filling so we don’t lose the original information about which values were missing.

This helps the model understand:

whether customer had credit access

how much credit they had

So we preserve information and keep data usable for modeling.

```python
data1["has_creditlimit"] = data1["creditlimit"].notnull().astype(int)
data1["creditlimit"].fillna(0, inplace=True)
```

```python
data1["has_cashlimit"] = data1["cashlimit"].notnull().astype(int)
data1["cashlimit"].fillna(0, inplace=True)
```

```python
data1["has_interest"] = data1["rateofinterest"].notnull().astype(int)
data1["rateofinterest"] = pd.to_numeric(data1["rateofinterest"], errors="coerce")
data1["rateofinterest"].fillna(0, inplace=True)
```

```python
data1["has_paymentfreq"] = data1["paymentfrequency"].notnull().astype(int)
data1["paymentfrequency"].fillna(0, inplace=True)
```

```python
data1["has_payment"] = data1["actualpaymentamount"].notnull().astype(int)
data1["actualpaymentamount"].fillna(0, inplace=True)
```
